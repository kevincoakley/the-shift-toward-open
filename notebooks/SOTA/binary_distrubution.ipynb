```python
import json
```

```python
labels_count = {
    "research_type": {0: 0, 1: 0},
    "result_outcome": {0: 0, 1: 0},
    "affiliation": {0: 0, 1: 0},
    "problem_description": {0: 0, 1: 0},
    "goal_objective": {0: 0, 1: 0},
    "research_method": {0: 0, 1: 0},
    "research_question": {0: 0, 1: 0},
    "hypothesis": {0: 0, 1: 0},
    "prediction": {0: 0, 1: 0},
    "contribution": {0: 0, 1: 0},
    "pseudocode": {0: 0, 1: 0},
    "open_source_code": {0: 0, 1: 0},
    "open_experiment_code": {0: 0, 1: 0},
    "train": {0: 0, 1: 0},
    "validation": {0: 0, 1: 0},
    "test": {0: 0, 1: 0},
    "results": {0: 0, 1: 0},
    "hardware_specification": {0: 0, 1: 0},
    "software_dependencies": {0: 0, 1: 0},
    "experiment_setup": {0: 0, 1: 0},
}
```

```python
# read the JSON file
# Use evaluations-fixed instead of evaluations-fixed-zeros to keep the original intent of SOTA
with open("../../evaluations-fixed.json", "r") as file:
    papers = json.load(file)

for paper in papers:
    for label in paper:
        if label in labels_count.keys():
            #print(paper[label])
            if paper[label] == -1:
                continue
            if paper[label] == 2:
                paper[label] = 1

            labels_count[label][paper[label]] += 1
```

```python

for label in labels_count:
    print(f"{label}: 0: {labels_count[label][0]} 1: {labels_count[label][1]} Total: {labels_count[label][0] + labels_count[label][1]}")
```
