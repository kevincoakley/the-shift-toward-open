```python
import os
import pandas as pd
import matplotlib.pyplot as plt

font_size = 20

# Dataset with updated numbers
data = {
    "Year": [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "NeurIPS": [411, 403, 569, 679, 1009, 1428, 1898, 2334, 2671, 3218, 4435],
    "IJCAI": [0, 569, 647, 664, 718, 846, 645, 586, 678, 639, 790],
    "ICML": [310, 270, 322, 434, 621, 773, 1084, 1183, 1233, 1828, 2610],
    "ICLR": [38, 31, 80, 245, 337, 502, 687, 859, 1094, 1573, 2260],
    "AAAI": [447, 651, 676, 645, 935, 1146, 1607, 1654, 1319, 1578, 2331],
}

# Convert to DataFrame
df = pd.DataFrame(data)
df.set_index("Year", inplace=True)

# Plot the stacked area chart
ax = df.plot(kind="area", stacked=True, figsize=(12, 6), color={
    "AAAI": "#f3b23e",
    "ICLR": "#e37137",
    "ICML": "#d62728",
    "IJCAI": "#e662bd",
    "NeurIPS": "#5dbaf7",
})

ax.set_xlim((2014, 2024))

# Add titles and labels
plt.title("Number of Published Papers per AI Conferences by Year", fontsize=font_size)
plt.xlabel("Year", fontsize=font_size - 2)
plt.ylabel("Number of Papers", fontsize=font_size - 2)

# Have the x-axis show every year
plt.xticks(df.index)

# Increase font size of tick labels
plt.xticks(fontsize=font_size - 4)
plt.yticks(fontsize=font_size - 4)

handles, labels = ax.get_legend_handles_labels()
# Add legend with larger font size
plt.legend(reversed(handles), reversed(labels), fontsize=font_size - 4)

# Adjust layout
plt.tight_layout()

if not os.path.exists("../../plots"):
    os.makedirs("../../plots")

plt.savefig(f"../../plots/papers_per_conference.pdf", dpi=600, bbox_inches='tight')

# Display the plot
plt.show()
```
