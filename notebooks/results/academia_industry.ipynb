```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
label_counts = {
    "academia_open_source_code": 0,
    "academia_open_datasets": 0,
    "industry_open_source_code": 0,
    "industry_open_datasets": 0,
    "academia_total": 0,
    "industry_total": 0,
    "total": 0
}
```

```python
exclude_theoretical = True

# Open llm_results.csv
csv_file_path = "../../results/llm_results.csv"

if not os.path.exists(csv_file_path):
    raise FileNotFoundError(f"CSV file not found: {csv_file_path}")

df = pd.read_csv(csv_file_path)
```

```python
len(df)
```

```python
if exclude_theoretical:
    # remove all of the rows where research_type is 1
    df = df[df["research_type"] != 1]
```

```python
len(df)
```

```python
label_results = {}

label_results = {
    2014: label_counts.copy(),
    2015: label_counts.copy(),
    2016: label_counts.copy(),
    2017: label_counts.copy(),
    2018: label_counts.copy(),
    2019: label_counts.copy(),
    2020: label_counts.copy(),
    2021: label_counts.copy(),
    2022: label_counts.copy(),
    2023: label_counts.copy(),
    2024: label_counts.copy(),
}
```

```python
df.head()
```

```python
for _, row in df.iterrows():
    year = row["year"]

    if row["affiliation"] == 0:
        if row["open_source_code"] == 1:
            label_results[year]["academia_open_source_code"] += 1
        if row["train"] == 1:
            label_results[year]["academia_open_datasets"] += 1

        label_results[year]["academia_total"] += 1

    elif row["affiliation"] == 1 or row["affiliation"] == 2:
        if row["open_source_code"] == 1:
            label_results[year]["industry_open_source_code"] += 1
        if row["train"] == 1:
            label_results[year]["industry_open_datasets"] += 1
    
        label_results[year]["industry_total"] += 1

    label_results[year]["total"] += 1
```

```python
# Convert label_results to percentages
for year in label_results:

    label_results[year]["academia_open_datasets"] = round(
        ((label_results[year]["academia_open_datasets"] / label_results[year]["academia_total"]) * 100)
    )
    label_results[year]["industry_open_datasets"] = round(
        ((label_results[year]["industry_open_datasets"] / label_results[year]["industry_total"]) * 100)
    )
    label_results[year]["academia_open_source_code"] = round(
        ((label_results[year]["academia_open_source_code"] / label_results[year]["academia_total"]) * 100)
    )
    label_results[year]["industry_open_source_code"] = round(
        ((label_results[year]["industry_open_source_code"] / label_results[year]["industry_total"]) * 100)
    )

    del label_results[year]["total"]

    del label_results[year]["academia_total"]
    del label_results[year]["industry_total"]
```

```python
years = list(label_results.keys())

# convert label_results to lists for plotting
academia_open_datasets = [label_results[year].get('academia_open_datasets', 0) for year in years]
industry_open_datasets = [label_results[year].get('industry_open_datasets', 0) for year in years]
academia_open_source_code = [label_results[year].get('academia_open_source_code', 0) for year in years]
industry_open_source_code = [label_results[year].get('industry_open_source_code', 0) for year in years]

plt.figure(figsize=(7, 6))
plt.plot(years, academia_open_datasets, color="blue", linestyle='-', label='Academia Open Datasets')
plt.plot(years, industry_open_datasets, color="blue", linestyle='--', label='Industry Open Datasets')
plt.plot(years, academia_open_source_code, color="red", linestyle='-', label='Academia Open Source Code')
plt.plot(years, industry_open_source_code, color="red", linestyle='--', label='Industry Open Source Code')

plt.title('Percentage of Papers with Open Source Code & \n Open Datasets Documented by Affiliation')
plt.xlabel('Year')
plt.ylabel('Percentage of Papers (Theoretical Excluded)')
plt.ylim(0, 100)

# Show y-axis ticks and labels on both left and right
plt.gca().yaxis.set_ticks_position('both')
plt.gca().tick_params(axis='y', labelright=True)

plt.gca().yaxis.set_major_locator(mticker.MultipleLocator(10))

plt.gca().grid(True, axis='x')

# Set x-axis ticks to show every year
plt.xticks(years)

plt.legend()
plt.grid()

plt.savefig("../../plots/academia_industry.pdf", dpi=600, bbox_inches='tight')

plt.show()
```
