```python
import os
import pandas as pd
```

```python
label_counts = {
    "public_code_public_data": 0,
    "private_code_private_data": 0,
    "total": 0,
}
```

```python
exclude_theoretical = True

# Open llm_results.csv
csv_file_path = "../../results/llm_results.csv"

if not os.path.exists(csv_file_path):
    raise FileNotFoundError(f"CSV file not found: {csv_file_path}")

df = pd.read_csv(csv_file_path)
```

```python
len(df)
```

```python
if exclude_theoretical:
    # remove all of the rows where research_type is 1
    df = df[df["research_type"] != 1]
```

```python
len(df)
```

```python
label_results = {}

label_results = {
    2014: label_counts.copy(),
    2015: label_counts.copy(),
    2016: label_counts.copy(),
    2017: label_counts.copy(),
    2018: label_counts.copy(),
    2019: label_counts.copy(),
    2020: label_counts.copy(),
    2021: label_counts.copy(),
    2022: label_counts.copy(),
    2023: label_counts.copy(),
    2024: label_counts.copy(),
}
```

```python
df.head()
```

```python
for _, row in df.iterrows():
    year = row["year"]

    if row["open_source_code"] == 1 and row["train"] == 1:
        label_results[year]["public_code_public_data"] += 1
    elif row["open_source_code"] == 0 and row["train"] == 0:
        label_results[year]["private_code_private_data"] += 1

    label_results[year]["total"] += 1
```

```python
# Convert label_results to percentages
for year in label_results:
    for label in label_results[year]:
        label_results[year][label] = round(
            ((label_results[year][label] / label_results[year]["total"]) * 100)
        )


    del label_results[year]["total"]
```

```python
# create a line plot of label_results
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

font_size = 18

years = list(label_results.keys())

public_code_public_data = [label_results[year]["public_code_public_data"] for year in years]
private_code_private_data = [label_results[year]["private_code_private_data"] for year in years]

plt.figure(figsize=(12, 6))
plt.plot(years, public_code_public_data, label='Both Open Code and Datasets', linewidth=3)
plt.plot(years, private_code_private_data, label='Neither Open Code nor Datasets', linewidth=3)
plt.title('Documentation of Open Code and Datasets', fontsize=font_size)
plt.xlabel('Year', fontsize=font_size - 2)
plt.ylabel('Percentage of Empirical Papers', fontsize=font_size - 2)
plt.ylim(0, 100)

# Show y-axis ticks and labels on both left and right
plt.gca().yaxis.set_ticks_position('both')
plt.gca().tick_params(axis='y', labelright=True)

plt.gca().yaxis.set_major_locator(mticker.MultipleLocator(10))

# Show x-axis ticks for every year
plt.gca().xaxis.set_major_locator(mticker.MultipleLocator(1))

plt.gca().grid(True, axis='x')

# Increase font size of tick labels
plt.xticks(fontsize=font_size - 4)
plt.yticks(fontsize=font_size - 4)

# Add legend with larger font size
plt.legend(fontsize=font_size - 4)

plt.grid()

plt.savefig("../../plots/both_open_code_data.pdf", dpi=600, bbox_inches="tight")
plt.show()
```
