```python

# Loop through each conference and year to process the results files
for conference in llm_results_files.keys():

    for year in llm_results_files[conference].keys():

        keys = []

        # Loop through each results file for the conference and year
        for results_file in llm_results_files[conference][year]:

            # Check if the results file exists
            if not os.path.exists(results_path + results_file):
                print(f"Results file {results_file} does not exist. Skipping.")
                continue
            
            # Read the results JSON file
            with open(results_path + results_file, 'r') as file:
                results = json.load(file)

            # Loop through the papers in the JSON data
            for key in results.keys():

                # Skip extra metadata keys
                if key in ignore_labels:
                    continue

                keys.append(key)

        try:
            df = pd.read_csv(f"../../paper_lists/downloaded_papers/{conference}_{year}.csv")
        except Exception as e:
            print(f"Error reading CSV file: {e}")
            
        if 'Filename' not in df.columns:
            print("The CSV file does not contain a 'Filename' column.")

        csv_files = df['Filename'].dropna().astype(str).to_list()

        # Remove the ".pdf" suffix from the filenames in the CSV
        csv_files = [f.replace('.pdf', '') for f in csv_files]

        sorted_keys = sorted(keys)
        sorted_csv_files = sorted(csv_files)

        # Print duplicate keys in results
        duplicates = [key for key in set(sorted_keys) if sorted_keys.count(key) > 1]
        if duplicates:
            print(f"\n{conference}_{year} Duplicate keys found in results:")
            for dup in duplicates:
                print(f"- {dup}")
        # Print duplicate keys in CSV
        csv_duplicates = [file for file in set(sorted_csv_files) if sorted_csv_files.count(file) > 1]
        if csv_duplicates:
            print(f"\n{conference}_{year} Duplicate keys found in CSV:")
            for dup in csv_duplicates:
                print(f"- {dup}")

        if len(sorted_keys) != len(sorted_csv_files):
            print(f"\n{conference}_{year} Mismatch in number of keys:")
            print(f"Keys in Results: {len(sorted_keys)}")
            print(f"Keys in CSV: {len(sorted_csv_files)}")

        #if len(missing_keys) > 0 or len(extra_keys) > 0:
        # Diff the two lists to find missing keys
        missing_keys = set(sorted_csv_files) - set(sorted_keys)
        if len(missing_keys) > 0:
            print(f"\n{conference}_{year} Files in CSV but NOT in Results:")
            for f in sorted(missing_keys):
                print(f"- {f}")

        # Diff the two lists to find extra keys
        extra_keys = set(sorted_keys) - set(sorted_csv_files)
        if len(extra_keys) > 0:
            print(f"\n{conference}_{year} Files in Results but NOT in CSV:")
            for f in sorted(extra_keys):
                print(f"- {f}")
```
